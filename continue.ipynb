```python
import pandas as pd
def convert_weather(cell):
    if cell == 'n.a':
        return 'Sunny'
    else:
        return cell
def convert_temp(cell):
    if cell <= 0 :
        return '14'
    else:
        return cell
df = pd.read_excel("C:\\Users\\lenovo\\Desktop\\trail.xlsx", 'Sheet1', converters = {'weather': convert_weather,
                                                                                    'temp': convert_temp })
df
```
